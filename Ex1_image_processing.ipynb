```python
import matplotlib.pyplot as plt
import numpy as np


fig = plt.figure(figsize=(14, 6))
plt.subplot(121)
plt.imshow(I1) # visualize image using imshow function
plt.title('First image')

plt.subplot(122)
plt.imshow(I2) # visualize image using imshow function
plt.title('Second image')
plt.show()
```

```python
IG1 = sk.color.rgb2gray(I1) #Convert I1 to grayscale

plt.imshow(IG1) # visualize image using imshow function
IG1.shape
```

```python
import matplotlib.pyplot as plt
fig,axes = plt.subplots(1,3,figsize=(12,7))
ax = axes.ravel() # convert axes to a 1d array

im1 = ax[0].imshow(IG1,cmap=plt.cm.gray) # visualize image using imshow function using a grayscale colormap
ax[0].set_title('grayscale image with grayscale colormap') # set figure title
fig.colorbar(im1,ax=ax[0],orientation='horizontal') # visualize colorbar

im2 = ax[1].imshow(IG1,cmap=plt.cm.jet) # visualize image using imshow function using a grayscale colormap
ax[1].set_title('grayscale image with jet colormap') # set figure title
fig.colorbar(im2,ax=ax[1],orientation='horizontal') # visualize colorbar

im2 = ax[2].imshow(IG1,cmap=plt.cm.winter) # visualize image using imshow function using a grayscale colormap
ax[2].set_title('grayscale image with winter colormap') # set figure title
fig.colorbar(im2,ax=ax[2],orientation='horizontal') # visualize colorbar

sk.io.show() #+ c'est blanc + c'est rouge, + c'est noir + c'est bleu
```

```python
# reshape the uint8 image as 1d array:
v = IG1.reshape(IG1.shape[0]*IG1.shape[1],1) #matrix convertion into vector
v=v*255
fig = plt.figure(figsize=(14, 6))
plt.hist(v, bins=256,density=False)
plt.show()
```
